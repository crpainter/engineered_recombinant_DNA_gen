```python
import glob, re
import pandas as pd
import pdb

import plotly.express as px
#from jupyter_dash import JupyterDash
#import dash_core_components as dcc
#import dash_html_components as html
#from dash.dependencies import Input, Output
import plotly.graph_objs as go


#Global variables:
df = pd.DataFrame()
fragments = ''
lines = 0
query = "MW289908.1"

def sort_files(query_name):
    numbers = []
    file_names = []
    for file in glob.glob("./results/analysis_sequences/windowed_sequences/" + query_name +"/*"):
        if ".out" in file:
            number = int(file.split('_')[-1].split('.')[:-1][0])
            numbers.append(number)
            numbers.sort()
    for i in numbers:
        file_names.append('fragment_' + str(i))
    return(file_names)
        
        
def readfiles(query_name):
    global df, fragments, lines
    fragments = sort_files(query_name)
    for fragment in fragments:
        with open ("./results/analysis_sequences/windowed_sequences/" + query_name + "/" + fragment + '.out') as myfile:
            dic = {}
            for line in myfile:  
                lines += 1
                if '>' in line or 'Identities' in line:
                    if '>' in line:
                        dic['Fragment'] = fragment
                        a = line.split()
                        dic['Agent'] = a[0][1:]
                    elif 'Identities' in line:
                        dic['Identities'] = str(line.split('Identities')[1].split()[1])
                        dic['Identities0'] = int(str(line.split('Identities')[1].split()[1]).split('/')[0])
                        dic['Identities1'] = int(str(line.split('Identities')[1].split()[1]).split('/')[1])
                        Identities_per = line.split('Identities')[1].split()[2][:-1]
                        dic['Identities_percentage']  = float(re.sub('[()]', '',Identities_per)[:-1])#Remove parehthesis 
                        dic['Gaps'] = line.split('Gaps')[1].split()[1]
                        Gaps_per = line.split('Gaps')[1].split()[2]
                        dic['Gaps_percentage'] = float(re.sub('[()]', '',Gaps_per)[:-1])#Remove parehthesis 
                elif 'No hits found' in line:
                    dic['Fragment'] = fragment
                    dic['Agent'] = 'No hits found'
                    dic['Identities'] = '0/0'
                    dic['Identities0'] = 0
                    dic['Identities1'] = 0
                    dic['Identities_percentage'] = 0
                    dic['Gaps'] = '0/0'
                    dic['Gaps_percentage'] = 0
                    
                else:
                    pass
                
                if 'Fragment' in dic and 'Identities0' in dic:  
                    df = df.append(dic, ignore_index = True )
                    dic = {}
                else:
                    pass
    print ("Total documents scanned: ", len(fragments))
    print ('Total lines scanned: ', lines)

readfiles(query)
no_hit_fragments = list(df[df['Agent'] == 'No hits found']['Fragment'])
no_hit_fragments.sort()

df
```

```python
df_max = pd.DataFrame()
for value in fragments:
    df.drop_duplicates(inplace=True)
    max_value = df[df['Fragment'] == value]['Identities0'].max()
    new_df = df[(df['Fragment'] == value) & (df['Identities0'] == max_value)]
    df_max = df_max.append(new_df, ignore_index=True)

df_max = df_max[['Fragment', 'Identities0']]
df_max.drop_duplicates(inplace=True)


title_font_color = "#ffffff"
title_font_family = 'Nunito Sans, sans-serif'
title_x=0.5
autosize=True
width=1600
height=350
paper_bgcolor = "white"
plot_bgcolor = "white"
template = 'plotly_dark'
axis_text_color = '#ab4edf'
fig = go.Figure()
fig.update_layout(
        title= '[Query Plasmid Name]',
        title_font_color=title_font_color,
        title_font_family= title_font_family, 
        title_x=title_x,
        autosize=autosize,
        width=width,
        height=height,
        paper_bgcolor=paper_bgcolor,
        plot_bgcolor=plot_bgcolor,
        template=template
        )

fig.add_trace(go.Bar(y=df_max.Identities0, x=df_max.Fragment))
fig.update_xaxes(visible=True, fixedrange=False, autorange = True,color=axis_text_color, rangeslider_visible=True, showspikes=True, spikethickness=2, spikecolor="#ffa500")
fig.update_yaxes(visible=True, fixedrange=False,  autorange = True, color=axis_text_color, showspikes=True, spikethickness=2, spikecolor="#ffa500")

config = dict({'scrollZoom': False, 'displayModeBar': True})

fig.show(config = config)
```
